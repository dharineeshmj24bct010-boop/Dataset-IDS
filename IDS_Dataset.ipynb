```python
def detect_attack(row):
    if row["Tot_Fwd_Pkts"] > 40:
        return "Possible DoS"
    if row["Tot_Bwd_Pkts"] > 50:
        return "Possible PortScan"
    if row["Flow_Bytes_s"] > 25000:
        return "Possible BruteForce"
    return "Normal"

df["predicted"] = df.apply(detect_attack, axis=1)
df
```
